```python
import numpy as np
import pandas as pd
import seaborn as sns,matplotlib.pyplot as plt
```

```python
msd=pd.read_csv("manager_survey_data.csv")
in_t=pd.read_csv("in_time.csv")
esd=pd.read_csv("employee_survey_data.csv")
data=pd.read_excel("data_dictionary.xlsx")
ou_t=pd.read_csv("out_time.csv")
gend=pd.read_csv("general_data.csv")
```

```python
in_t=in_t.replace(np.nan,0)
in_t.iloc[:, 1:] = in_t.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
in_t.head()
```

```python
ou_t=ou_t.replace(np.nan,0)
ou_t.iloc[:, 1:] = ou_t.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
ou_t.head()
```

```python
in_t=in_t.append(ou_t)
in_t.head()
```

```python
in_t=in_t.diff(periods=4410)
in_t=in_t.iloc[4410:]
in_t.reset_index(inplace=True)
in_t.head()
```

```python
in_t.drop(columns=['index','Unnamed: 0'],axis=1,inplace=True)
in_t.head()
```

```python
in_t.shape
```

```python
in_t.drop(['2015-01-01', '2015-01-14','2015-01-26','2015-03-05',
             '2015-05-01','2015-07-17','2015-09-17','2015-10-02',
              '2015-11-09','2015-11-10','2015-11-11','2015-12-25'
             ], axis = 1,inplace=True)
```

```python
in_t['Actual Time']=in_t.mean(axis=1)
in_t['hrs']=in_t['Actual Time']/np.timedelta64(1, 'h')
```

```python
in_t.reset_index(inplace=True)
in_t.head()
```

```python
in_t.drop(in_t.columns.difference(['index','hrs']),1,inplace=True)
```

```python
in_t.rename(columns={'index': 'EmployeeID'},inplace=True)
in_t.head()
```

# Combining Data

```python
df_1 = pd.merge(esd, gend, how='inner', on='EmployeeID')
hr = pd.merge(msd, df_1, how='inner', on='EmployeeID')
hr = pd.merge(in_t, hr, how='inner', on='EmployeeID')
hr.head()
```

```python
hr['JobLevel']=hr['JobLevel'].astype('object')
```

```python
hr['Education'] = hr['Education'].replace({ 1 : 'Below College', 2: 'College',3: 'Bachelor',4: 'Master',5 : 'Doctor'})
hr['EnvironmentSatisfaction'] = hr['EnvironmentSatisfaction'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
hr['JobInvolvement'] = hr['JobInvolvement'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
hr['JobSatisfaction'] = hr['JobSatisfaction'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
#hr['RelationshipSatisfaction'] = hr['RelationshipSatisfaction'].replace({ 1 : 'Low', 2: 'Medium',
#                   3: 'High',4: 'Very High'})
hr['PerformanceRating'] = hr['PerformanceRating'].replace({ 1 : 'Low', 2: 'Good',3: 'Excellent',4: 'Outstanding'})
hr['WorkLifeBalance'] = hr['WorkLifeBalance'].replace({ 1 : 'Bad', 2: 'Good',3: 'Better',4: 'Best'})
```

```python
hr.drop(['EmployeeID', 'EmployeeCount','StandardHours','Over18'], axis = 1,inplace=True)
```

```python
sns.countplot(x='EnvironmentSatisfaction',data=hr);
```
